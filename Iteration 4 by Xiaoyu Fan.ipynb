```python
print(dataframe['Burned Level'].value_counts())

# Separate majority and minority classes
df_majority = dataframe[dataframe['Burned Level']==0]
df_minority = dataframe[dataframe['Burned Level']==2]
 
# Upsample minority class
from sklearn.utils import resample
df_minority_upsampled = resample(df_minority, 
                                 replace=True,     
                                 n_samples=247,    
                                 random_state=123) 
# Combine majority class with upsampled minority class
df_upsampled_High = pandas.concat([df_majority, df_minority_upsampled])

df_minority = dataframe[dataframe['Burned Level']==1]
 
# Upsample minority class
df_minority_upsampled = resample(df_minority, 
                                 replace=True,     
                                 n_samples=247,    
                                 random_state=123) 
# Combine majority class with upsampled minority class
df_upsampled = pandas.concat([df_majority, df_minority_upsampled])
df_upsampled_None=df_upsampled[~df_upsampled['Burned Level'].isin([0])]
# Display new class counts
dataframe = df_upsampled_None.append(df_upsampled_High,ignore_index=True)
print(dataframe['Burned Level'].value_counts())
```
